```python
# Total complaints of top 20 issues.
figsize(10, 5)
IssueWise=d['Issue'].value_counts()[:20]
IssueWise.reindex(index=IssueWise.index[::-1]).plot(kind="barh",title="Total complaints of top 20 issues")
```
